```python
'''since it's outfield player dataframe, we will not need any GK stats. so let's 
drop them and also, since we subset this dataframe from the original, might 
aswell reset the index'''
df_outfield.drop(columns =['id','GKHandling','GKDiving','GKKicking','GKPositioning','GKReflexes']
                 ,inplace = True)
df_outfield.reset_index(drop = True, inplace = True)
df_outfield.info()
```

```python
#check the height and weight column, and see how we'll tackle this column
df_outfield[['Height','Weight']].head()
```

```python
#to change it to cm, we can use split like this
df_outfield['Height'].str.split("'")[0]
```

```python
'''now, since the height column is not in the right datatype, we'll change it to 
cm using a function'''
def ConvertHeight(x):
  #The column values should look something like this 5'7
   ht_ = x.split("'")
   ft_ = round(float((ht_[0])),2) * 12 * 2.54
   in_ = round(float((ht_[1])),2) * 2.54
   return (ft_ + in_)

df_outfield['Height'] = df_outfield['Height'].apply(lambda x : ConvertHeight(x))
df_outfield['Height'].head()
```

```python
#now on to the next one
print(df_outfield['Weight'].head())
```

```python
#check whether each column ends with lbs
df_outfield['Weight'].str.contains('lbs').sum()
#since this is the case, then we'll use string manipulation
```

```python
#let's convert it to Kg
df_outfield['Weight'] = df_outfield['Weight'].str[:-3].astype(float)
df_outfield['Weight'] = round(df_outfield['Weight'] * 0.453592,2)
df_outfield['Weight'].head()
```

```python
df_outfield.info()
#looks like both weight and height column are already in the correct dtype
```

```python
#another columns that needs to change are Values, Wage and Release Clause
df_outfield[['Value','Wage','Release Clause']].head()
```

```python
'''
since for value column ends with either M or K, we need to check if it's
consistent all the way 
'''
nominals = df_outfield['Value'].str[-1]
nominals.value_counts()
```

```python
#looks like that column ends with either M K or 0
df_outfield['Value'].str.contains('M|K|0').sum()
```

```python
df_outfield[df_outfield['Value'].str[-1]=='0']
'''
players with value ends with 0 means they have no value
'''
```

```python
'''
just like before, we'll make a function to convert it and apply it to each 
column
'''
def EuroConvert(x):
  # format : Eurosign .... K/M/0
  l = len(x)
  end = x[-1]
  if end == 'M' :
    value = float (x[1:l-1]) * 1000000
  elif end == 'K' :
    value = value = float (x[1:l-1]) * 1000
  else :
    value = 0
  return value

df_outfield['Value'] = df_outfield['Value'].apply(lambda x: EuroConvert(x))
```

```python
df_outfield.info()
```

```python
#another columns that needs to change are Wage and Release Clause
df_outfield[['Wage','Release Clause']].head()
```

```python
'''
since both Wage and Release Clause column are in the same format, we can use 
the previous function
'''
df_outfield['Wage'] = df_outfield['Wage'].apply(lambda x : EuroConvert(x))
df_outfield['Release Clause'].fillna("None", inplace = True)
df_outfield['Release Clause'] = df_outfield['Release Clause'].apply(lambda x : EuroConvert(x))
```

```python
df_outfield['Attacking Work Rate'] = df_outfield['Work Rate'].str.split('/', expand = True)[0]
df_outfield['Defensive Work Rate'] = df_outfield['Work Rate'].str.split('/', expand = True)[1]
df_outfield[['Work Rate','Attacking Work Rate','Defensive Work Rate']].tail()
```

```python
'''
now that these values have been changed, some columns will be drop and for Body
Type column, upon further search, i can't seem to find what some of the players
body type is, so i'll just drop them
'''
df_outfield.drop(columns = ['Special','Work Rate','Joined','Body Type'], inplace = True)
df_outfield.info()
```

```python
'''
finally, let's change the positions rating to integer, note : the '+' sign is 
present since from the start of the game, the players overtime the players improve
in this dataset, we'll use the ratings overtime
'''
df_outfield[['LS', 'ST', 'RS', 'LW', 'LF', 'CF', 'RF', 'RW', 'LAM', 'CAM',
       'RAM', 'LM', 'LCM', 'CM', 'RCM', 'RM', 'LWB', 'LDM', 'CDM', 'RDM',
       'RWB', 'LB', 'LCB', 'CB', 'RCB', 'RB']].head()
```

```python
def RatingOvertime (x): 
  Rt = x.split('+')
  return int(Rt[0]) + int(Rt[1])

iterate = ['LS', 'ST', 'RS', 'LW', 'LF', 'CF', 'RF', 'RW', 'LAM', 'CAM',
       'RAM', 'LM', 'LCM', 'CM', 'RCM', 'RM', 'LWB', 'LDM', 'CDM', 'RDM',
       'RWB', 'LB', 'LCB', 'CB', 'RCB', 'RB']
for i in iterate : 
  df_outfield[i] = df_outfield[i].apply(lambda x : RatingOvertime(x))

df_outfield[iterate].head()
```
